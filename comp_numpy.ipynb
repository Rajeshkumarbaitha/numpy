```python
# 6 array stacking and splitting
```

```python
a= np.array([1,2,3,4])
b= np.array([5,6,7,8])
print(np.vstack((a,b)))
print(np.hstack((a,b)))
```

```python
#splitting
c= np.array([[1,2,3],[4,5,6]])
print (c)
```

```python

hsplit=np.hsplit(c,3)
for s in hsplit:
    print(s)
```

```python

vsplit=np.vsplit(c,2)
for s in vsplit:
    print(s)
```
